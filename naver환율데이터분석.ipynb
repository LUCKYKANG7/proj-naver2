## 네이버 환율정보  엑셀 저장

```python
#html은 테그를 사용하는데 브라우저가 인터프리터의 역할을 합니다. 
#브라우저는 같은 이름의 객체가 2개 이상이면 자동으로 배열로 변경 해줍니다. 
df = tables[0] 
```

```python
# 오늘 날짜 시간 정보를 지원하는 라이브러리   
from datetime import datetime
datetime.now()
```

```python
currtime = datetime.now()
df.to_excel("네이버 환율 정보_"+ currtime.strftime("%Y-%m-%d") + ".xlsx", index=True)
```
